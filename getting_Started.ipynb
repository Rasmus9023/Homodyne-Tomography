```python
import numpy as np
import qutip as q
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
```

### 2d visualization of the wigner function of a 2-photon fock state (2nd excited state)

```python
xvec = np.linspace(-4,4,121)
N_ph = 2
vac = q.basis(100,N_ph)
wig = q.wigner(vac, xvec, xvec)
```

```python
cs = plt.contourf(xvec,xvec,wig,100)
cbar = plt.colorbar(cs)

plt.xlabel('x values')
plt.ylabel('p values')
plt.show()
```
